# 1. Sales Analysis:

```python
segment_sales=df.groupby("customer_segment")["sales"].sum().to_dict()
plt.pie(segment_sales.values(),labels=segment_sales.keys(),autopct=lambda p: '{:.0f} ({:.1f}%)'.format(p * sum(segment_sales.values()) / 100, p))
plt.title("Sales distribution as per customer segment")
plt.show()
```

```python
df["year_month"] = pd.to_datetime(df["order_date"],format="%Y-%m").dt.to_period('M')
df["year_month"] = df["year_month"].astype(str)
df["year_month"] = df["year_month"].apply(extract_date)
df_sorted = df.sort_values(by='order_date', ascending=True)

sale_trend = df.groupby("year_month")["sales"].sum().to_dict()

# Visualizing with plot
plt.figure(figsize=(12,6))
sns.lineplot(x = list(sale_trend.keys()) , y = list(sale_trend.values()) , marker='o')
plt.xlabel("Dates as per month and year")
plt.ylabel("Sales")
plt.title("Sales trend overtime")
plt.grid(True, axis='both')
plt.xticks(rotation=90)
plt.show()
```

```python
df["month"] = pd.to_datetime(df["order_date"],format="%Y-%m").dt.month
df_sorted = df.sort_values(by='order_date', ascending=True)

sale_trend_month = df.groupby("month")["sales"].sum().to_dict()

# Visualizing with plot
plt.figure(figsize=(12,6))
sns.lineplot(x = list(sale_trend_month.keys()) , y = list(sale_trend_month.values()) , marker='o')
plt.xlabel("Months of each year")
plt.ylabel("Sales")
plt.title("Sales trend overtime accorindin to month")
plt.grid(True, axis='both')
plt.xticks(rotation=90)
plt.show()
```

    Overall Sales started started to decline from July 2017 , by analysing both charts we can see trend started to decline between June to December month.There is seasonal factor impacting Sales.

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'
    else:
        return 'Unknown'

df["seasons"] = df["month"].apply(get_season)

seasonal_trend = df.groupby("seasons")["sales"].sum().sort_values(ascending=False).to_dict()

plt.figure(figsize=(12,6))
plt.bar(seasonal_trend.keys(), seasonal_trend.values(), color='lightblue', label='Sales')
plt.xlabel("Seasons")
plt.ylabel("Sales")
plt.title("Sales trend with seasons")
plt.show()
```

    Sales declined in Fall and Winter seasons.By Analyzing all trends chart and season chart we can seen season 'Fall' , Sep, Oct and Nov months and Year 2017 impacting Sales to downward.

# 6. Discount Impact:

```python
# Visualizing relation between discount rate & profit ratio
plt.figure(figsize=(10, 6))
plt.scatter(df["order_item_discount_rate"], df["order_item_profit_ratio"], alpha=0.7)
plt.title("Discount rate vs Profit Ratio", fontsize=14)
plt.xlabel("Discount Rate", fontsize=12)
plt.ylabel("Profit Ration", fontsize=12)
plt.grid(True)
plt.show()
```

The scatter plot shows that higher discount rates do not correspond to consistent positive profit ratios. In fact, there is a wide spread, with a large portion of orders showing negative profit ratios as the discount rate increases.
The profit ratio seems more volatile and scattered across different discount rates, with a significant number of transactions showing losses (negative profit ratios), especially as discounts increase.
This indicates that deeper discounts are not always beneficial for maintaining profitability, and businesses should be cautious when applying higher discount rates.

```python
# Visualizing relation between discount amount & profit amount
plt.figure(figsize=(10, 6))
plt.scatter(df["order_item_discount"], df["order_profit_per_order"], alpha=0.7)
plt.title("Discount Amount vs Profit Amount", fontsize=14)
plt.xlabel("Discount Amount", fontsize=12)
plt.ylabel("Profit Amount", fontsize=12)
plt.grid(True)
plt.show()
```

There is a broad range of discount amounts, but for the majority of orders, profit tends to stay around zero, regardless of the discount amount.
While some higher discount values (up to 500) seem to be associated with significant losses (negative profits), there are also outliers where higher discounts coincide with positive profits.
The trend suggests that higher discounts do not necessarily guarantee higher profits and, in fact, often lead to either neutral or negative profit outcomes.

Need to optimization for each quantity , needs to plan quantity and discount related strategies to improve profit and reducing loss.

The analysis reveals that increasing order quantities does not necessarily lead to higher profitability, with losses of up to 1000 occurring at various quantity levels. Most of the profits come from single-quantity orders, indicating that the current quantity scaling approach is inefficient.

To improve overall profitability, optimization is required for each quantity bracket. Developing strategic plans around quantity-based discounts, bundling offers, and improving operational efficiencies will help reduce losses and drive higher profits per order.
